# Data Reading

```python
df_sales = spark.read.format('csv')\
  .option('header', True)\
  .option('inferSchema', True)\
  .load('abfss://source@datalakejubin.dfs.core.windows.net')
```

```python
df_sales.display()
```

### Pyspark Transformations

```python
df_sales.withColumn('Item_Type', split(col('Item_Type'),' ')).display()
```

```python
df_sales.withColumn('flag', lit(var)).display() ## lit is for constants
```

```python
df_sales.withColumn('Item_Visibility', col('Item_Visibility').cast(StringType())).display()
```
